Purpose :
To predict the underwater ROCK or Explosive MINE .

Description :
The dataset contains a set of 208 records under 60 attributes - Data consist of sonar signals either bounced off a metal cylinder or bounced off a roughly cylindrical rock and Class(R - Rock , M - Mine).

STEP 1 : Load all the necessary libraries

```
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
import pandas as pd
import numpy as np
pd.set_option('display.max_rows',None)
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```
df = pd.read_csv("sonar.csv",header=None)
```

```
df
```


STEP 2 : Data Preparation

```
sonar = df.sample(frac = 1) 
```

```
sonar = sonar.values
```

```
sonar_xtrain = sonar[:124 , :60].astype('float')
sonar_xtrain.shape
```

```
sonar_ytrain = sonar[:124 , 60]
sonar_ytrain.shape
```

```
sonar_xtest = sonar[124: , :60].astype('float')
sonar_xtest.shape
```

```
sonar_ytest = sonar[124: , 60]
sonar_ytest.shape
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_ytrain = le.fit_transform(sonar_ytrain)
encoded_ytrain
```

Rock = 1 Mine = 0

```
encoded_ytest = le.fit_transform(sonar_ytest)
encoded_ytest
```



STEP 3 : Model Architecture 

```
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(60,)),     # Convert in 1D
    tf.keras.layers.Dense(60 , activation='relu'),   #  max(x,0)
    tf.keras.layers.Dense(1, activation = 'sigmoid')   # In binary we use sigmoid
])
```

STEP 4 : Compilation Step 

```
model.compile(optimizer='adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])
```

```
model.fit(sonar_xtrain , encoded_ytrain , epochs = 300)
```

STEP 5 : Evaluation

```
model.evaluate(sonar_xtest , encoded_ytest)
```

```
loss , acc = model.evaluate(sonar_xtest , encoded_ytest , verbose=2)
```
